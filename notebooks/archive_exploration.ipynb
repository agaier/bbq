### Load Sample Grid Archive

```python
data_path = 'example_archive'
folder = f'{data_path}/'

p = load_config(get_config_files(folder))['archive']
fit_grid = np.load(folder+'/_fit.npy')
desc_grid = np.load(folder+'/_desc.npy')
```

### View Objective in each bin

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
from plotting import set_map_grid

def view_map(Z, p, ax=None, bin_ticks=False):
    if ax is None:
        fig,ax = plt.subplots(figsize=(4,4),dpi=150)    
    with plt.style.context('me_grid'):        
        im = ax.imshow(Z, cmap='YlGnBu')

        # Colorbar
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.1)
        cb = plt.colorbar(im, cax=cax)
        for t in cb.ax.get_yticklabels():
            t.set_horizontalalignment('right')   
            t.set_x(3.0)

        # Grid
        set_map_grid(ax, Z, bin_ticks=bin_ticks, **p)
        
        ax.set(xlabel = p['desc_labels'][0], ylabel= p['desc_labels'][1])
                
    return ax

view_map(fit_grid, p)
#view_map(fit_grid, p, bin_ticks=True)
```

```python
fig,ax = plt.subplots(ncols=2, figsize=(9,4),dpi=150)    
ax[0] = view_map(fit_grid, p, ax=ax[0])
ax[1] = view_map(desc_grid[:,:,0], p, ax=ax[1])

```
